```python
import pandas as pd
df = pd.read_csv("sleeptime_prediction_dataset.csv")
df = df[df['SleepTime'] < 10]
df = df[df['SleepTime'] > 2.5]
df=df.drop(columns=['CaffeineIntake'])
```

```python
adv_df_output = df['SleepTime']
adv_df_input = df.drop(columns=['SleepTime'])

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(adv_df_input, adv_df_output, test_size = 0.3, random_state=1)

# I decided to only use a test size of 0.3 as I did not want to overfit or underfit the data.


from sklearn.linear_model import LinearRegression


lr = LinearRegression() 


lr.fit(x_train, y_train) 
```

```python
print('--> Linear influence from fixed acidity :       ',lr.coef_[0])  
print('--> Linear influence from volatile acidity :   ',lr.coef_[1])  
print('--> Linear influence from citric acid :        ',lr.coef_[2])  
print('--> Linear influence from residual sugar :      ',lr.coef_[3])  
print('--> Linear influence from chlorides :          ',lr.coef_[4])  
```

```python
import pickle

# Save to file
with open('linear_model.pkl', 'wb') as file:
    pickle.dump(lr, file)

print("Model saved as 'linear_model.pkl'")

```
